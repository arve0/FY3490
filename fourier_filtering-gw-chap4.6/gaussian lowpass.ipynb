Gaussian lowpass filter, page 276.

$ H(u,v) = e^{\frac{-D^2(u,v)}{2D_0}} $

```python
%run ../common.ipynb
from __future__ import division
image = imread('../MP.tiff')
F = fft2(image)
```

```python
D = 100
y,x = F.shape
X,Y = meshgrid(range(x), range(y))
H = numpy.exp(-((X-x/2)**2 + (Y-y/2)**2)/2/D**2)
gimshow(H)
```

```python
G = fftshift(F)*H
G = ifftshift(G)
g = ifft2(G).real
gimshow(g)
```

```python
imsave('lp-gaussian.tif', g.astype(np.uint8))
```
