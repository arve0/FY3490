## Ideal highpass fourier filter.

Page 281

```python
%run ../common.ipynb
image = imread('../MP.tiff')
F = fft2(image)
```

```python
y,x = F.shape
X,Y = meshgrid(range(x), range(y))
mask = sqrt((X-x/2)**2 + (Y-x/2)**2) < 100
F = fftshift(F)
F[mask] = 0
F = ifftshift(F)
f = ifft2(F).real
```

```python
gimshow(f)
```

```python
f[f < 0] = 0 # cut off negative values
imsave('hp-ideal.tif', f.astype(np.uint8))
```
