```python
# power-law spatial filter
from matplotlib import pyplot as plt
from matplotlib import image
import numpy as np
%matplotlib inline
img = image.imread('MP.tif')[...,0]
plt.imshow(img, cmap='gray')
```

```python
gamma = 0.5
gamma_img = img**gamma
plt.imshow(gamma_img, cmap='gray')
```

```python
gamma = 2
gamma_img = img**gamma
plt.imshow(gamma_img, cmap='gray')
```
