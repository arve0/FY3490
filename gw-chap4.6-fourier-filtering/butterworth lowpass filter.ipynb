Butterworth lowpass filter, page 273.

$ H(u,v) = \frac{1}{1+[D(u,v)/D_0]^{2n}} $

```python
%run ../common.ipynb
image = imread('../MP.tiff')
F = fft2(image)
```

```python
n = 2
D = 100
y,x = F.shape
X,Y = meshgrid(range(x), range(y))
H = 1/(1 + (sqrt((X-x/2)**2 + (Y-y/2)**2)/D)**(2*n))
gimshow(H)
```

```python
G = fftshift(F)*H
G = ifftshift(G)
g = ifft2(G).real
gimshow(g)
```

```python
imsave('butterworth.tif', g.astype(np.uint8))
```
