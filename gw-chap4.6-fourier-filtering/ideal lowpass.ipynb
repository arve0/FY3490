```python
image = imread('../MP.tiff')
gimshow(image)
histshow(image)
```

```python
F = fft2(image)
```

```python
F.shape
```

```python
x,y = meshgrid(range(F.shape[1]), range(F.shape[0]))
mask = sqrt((x-F.shape[0]/2)**2 + (y-F.shape[1]/2)**2) > 100
F = fftshift(F)
F[mask] = 0
F = ifftshift(F)
```

```python
f = ifft2(F).real
```

```python
f
```

```python
gimshow(f)
histshow(f.astype(np.uint8))
```

```python
imsave('ideal-lp.tif', f.astype(np.uint8))
```
